```python
import requests
import os
from google.cloud import storage
from google.oauth2 import service_account
import pandas as pd
import io
```

```python
address = f"https://www.googleapis.com/youtube/v3/search?q=the+needle+drop&key={os.environ.get('YOUTUBE_API_KEY')}&part=snippet"
# address = f"https://www.googleapis.com/youtube/v3/search?q=neo4j&key={os.environ.get('YOUTUBE_API_KEY')}&part=snippet"
# address = f"https://www.googleapis.com/youtube/v3/channels?id=UCvze3hU6OZBkB1vkhH2lH9Q&key={os.environ.get('YOUTUBE_API_KEY')}&part=contentDetails"


req = requests.get(address)
data = req.json()
```

```python
print(address)
```

```python
channel_id = data['items'][0]['snippet']['channelId']
```

```python
address = f"https://www.googleapis.com/youtube/v3/channels?id={channel_id}&key={os.environ.get('YOUTUBE_API_KEY')}&part=contentDetails"
req = requests.get(address)
data = req.json()
```

```python
uploads_id = data['items'][0]['contentDetails']['relatedPlaylists']['uploads']
```

```python
address = f"https://www.googleapis.com/youtube/v3/playlistItems?playlistId={uploads_id}&key={os.environ.get('YOUTUBE_API_KEY')}&part=snippet&maxResults=50"
req = requests.get(address)
data = req.json()
```

```python
credentials = service_account.Credentials.from_service_account_file(
        os.environ.get('GCP_SERVICE_ACCOUNT_KEY_PATH')
    )  
client = storage.Client(credentials=credentials)
bucket = client.get_bucket(os.environ.get("GCP_BUCKET_NAME"))
```

```python
videos_temp = bucket.get_blob("youtube/youtube_video_ids.csv")
videos_temp = videos_temp.download_as_string()
videos_list = pd.read_csv(io.BytesIO(videos_temp))['YouTube_video_ids'].to_list()
```
